```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
min_util = [0.00001, 0.00005, 0.0001, 0.0005, 0.001]
data = {
    "BMS" : {
        "min_corr" : 1.0,
        "time": [10, 8, 6, 5, 2]
    },
    "Chess" : {
        "min_corr" : 1.0,
        "time": [10, 8, 6, 5, 2]
    },
    "ECOM" : {
        "min_corr" : 1.0,
        "time": [10, 8, 6, 5, 2]
    },
    "Foodmart" : {
        "min_corr" : 1.0,
        "time": [10, 8, 6, 5, 2]
    },
    "Mushroom" : {
        "min_corr" : 1.0,
        "time": [10, 8, 6, 5, 2]
    }
}
```

```python
for name, data in data.items():
    print(name)
    plt.figure(figsize=(16,9))
    plt.plot(min_util, data['time'], marker='o')
    plt.xlabel('Minimum utility threshold', fontsize=16)
    plt.ylabel("Runtime (in seconds)", fontsize=16)
    plt.title(f'Minimum utility Vs Time - {name}', fontsize=20, color=(5/255,59/255,117/255))
    plt.grid()
    plt.legend()
    plt.savefig(f'./graphs/{name}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
